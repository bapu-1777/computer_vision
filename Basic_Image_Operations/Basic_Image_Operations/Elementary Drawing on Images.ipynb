## Import Libraries

```python
# Import libraries
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

## Read an image

```python
# Load the image
image = cv2.imread("Images/dog.jpg")

# Display the original image
plt.imshow(image[:,:,::-1])
```

## Draw a Line

We can use cv2.line function for this task.

Function Syntax: cv2.line(img, pt1, pt2, color[, thickness[, lineType[, shift]]]) ->	img

Parameters:

    img	Image.
    pt1	First point of the line segment.
    pt2	Second point of the line segment.
    color	Line color.
    thickness	Line thickness.
    lineType	Type of the line. See LineTypes.
    shift	Number of fractional bits in the point coordinates.



```python
# Draw a line
imageLine = image.copy()

# The line starts from (120,80) and ends at (270,80)
# The color of the line is RED (Recall that OpenCV uses BGR format)
# Thickness of line is 5px
# Linetype is cv2.LINE_AA

cv2.line(imageLine, (120, 80), (270, 80), (0, 255, 0), thickness=5, lineType=cv2.LINE_AA);

# Display the image
plt.imshow(imageLine[:,:,::-1])
```

```python
cv2.line(imageLine, (120, 200), (270, 200), (0, 255, 0), thickness=5, lineType=cv2.LINE_AA);
cv2.line(imageLine, (120, 80), (120, 200), (0, 255, 0), thickness=5, lineType=cv2.LINE_AA);
cv2.line(imageLine, (270, 80), (270, 200), (0, 255, 0), thickness=5, lineType=cv2.LINE_AA);
# Display the image
plt.imshow(imageLine[:,:,::-1])
```

## Drawing a rectange

To draw a rectangle, you need top-left corner and bottom-right corner of rectangle. This time we will draw a red rectangle on the dog head.

Function Syntax: cv2.rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]]) -> img

The mandatory arguments are as follows.

    img: Image on which the rectangle is to be drawn.
    pt1: Vertex of the rectangle. Usually we use the top-left vertex here.
    pt2: Vertex of the rectangle opposite to pt1. Usually we use the bottom-right vertex here.
    color: Rectangle color
    
    Next, there are some optional arguments:
    thickness: Thickness of the circle outline (if positive). If a negative value is supplied for this argument, it will result in a filled circle.
    lineType: Type of the circle boundary. This is exact same as lineType argument in cv2.line

```python
# Draw a rectangle
imageRec = image.copy()

# The top corner is at (120,80) and bottom corner at (270,200)
# The color of the line is RED (Recall that OpenCV uses BGR format)
# Thickness of line is 5px
# Linetype is cv2.LINE_AA

cv2.rectangle(imageRec, (120, 80), (270, 200), (255, 0, 0), thickness=10, lineType=cv2.LINE_AA);

# Display the image
plt.imshow(imageRec[:,:,::-1])
```

## Draw a Circle

We can use cv2.circle function to draw a circle on an image.

Function Syntax: cv2.circle(img, center, radius, color[, thickness[, lineType[, shift]]]) -> img

The mandatory arguments:

    img: Image where the circle is drawn.
    center: Center of the circle.
    radius: Radius of the circle.
    color: Circle color
    
    The optional arguments are same as rectangles

```python
# Draw a Circle
imageCircle = image.copy()

# The center is at (200,150) and the radius is 75
# The color of the line is BLUE (Recall that OpenCV uses BGR format)
# Thickness of line is 5px
# Linetype is cv2.LINE_AA

cv2.circle(imageCircle, (200, 150), 100, (255, 0, 0), thickness=5, lineType=cv2.LINE_AA);

# Display the image
plt.imshow(imageCircle[:,:,::-1])
```

## Drawing an Ellipse

Function Syntax: cv2.ellipse(img, center, axes, angle, startAngle, endAngle, color[, thickness[, lineType[, shift]]]) -> img

The mandatory arguments are as follows.

    img: Image on which the ellipse is to be drawn.
    center: Center of the ellipse.
    axes: radius of the ellipse major and minor axes.
    angle: Ellipse rotation angle in degrees.
    startAngle: Starting angle of the elliptic arc in degrees.
    endAngle: Ending angle of the elliptic arc in degrees.
    color: Ellipse line color
    
    The optional arguments are same as above

```python
# Draw an Ellipse
imageEli = image.copy()

# The center is at (200,150) and the radius is 75
# The color of the line is BLUE (Recall that OpenCV uses BGR format)
# Thickness of line is 5px
# Linetype is cv2.LINE_AA

cv2.ellipse(imageEli, (200, 150), (125,75), 0, 0, 360, (255, 0, 0), thickness=5, lineType=cv2.LINE_AA);
cv2.ellipse(imageEli, (200, 150), (125,75), 90, 0, 360, (0, 0, 255), thickness=5, lineType=cv2.LINE_AA);

# Display the image
plt.imshow(imageEli[:,:,::-1])
```

## Adding Texts to Images

Function Syntax: cv2.putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]]) -> img

The mandatory arguments:

    img: Image on which the text has to be written.
    text: Text string to be written.
    org: Bottom-left corner of the text string in the image.
    fontFace: Font type
    fontScale: Font scale factor that is multiplied by the font-specific base size.
    color: Font color
    
    The optional arguments that we are same as before.

```python
imageText = image.copy()

cv2.putText(imageText,'Hello World!!',(10,50), cv2.FONT_HERSHEY_SIMPLEX, 2,(255,255,255),2,cv2.LINE_AA);

# Display Image with Text
plt.imshow(imageText[:,:,::-1])
```
